# Best Route to Downtown
### Based on Different Metrics
1. The route with the least stops

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from CSV file
file_path = 'wprdc_stop_data.csv'
df = pd.read_csv(file_path)

# Group by route and count the number of stops
route_stops = df.groupby('route_name').size().reset_index(name='stop_id')

# Sort routes by stop count in descending order
route_stops = route_stops.sort_values(by='stop_id', ascending=False)

# Select top 10 routes with the most stops
top_routes = route_stops.head(10)

# Create a bar plot to display routes with the most stops
plt.figure(figsize=(10, 6))
sns.barplot(x='route_name', y='stop_id', data=top_routes)
plt.title('Top 10 Routes with Most Stops')
plt.xlabel('Route Name')
plt.ylabel('Stop Count')
plt.xticks(rotation=45)
plt.show()
```

# Route With least monthly average riders

```python
# Read data from CSV file
file_path = "average_ridership.csv"
df = pd.read_csv(file_path)

# Group by route and calculate average ridership
df_grouped = df.groupby("route").mean()

# Create a bar plot to represent average ridership per route
df_grouped.plot(kind="bar", legend=None)
plt.xlabel("Route")
plt.ylabel("Average Ridership")
plt.title("Average Ridership per Route")
plt.show()
```

### Limit to the bottom 25 per amount of avg riders

```python
# Read data from CSV file
file_path = "average_ridership.csv"
avg = pd.read_csv(file_path)

# Group by route and calculate average ridership
df_grouped_avg = avg.groupby("route").mean()

# Sort by average ridership in ascending order
df_grouped_sorted = df_grouped_avg.sort_values(by="avg_riders", ascending=True)

# Select top 25 routes with least ridership
df_top_25 = df_grouped_sorted.head(25)

# Create a bar plot to represent average ridership per route
df_top_25.plot(kind="bar", legend=None)
plt.xlabel("Route")
plt.ylabel("Average Ridership")
plt.title("Top 25 Routes with Least Average Ridership")
plt.show()
```

# Monthly On Time Performance by Route

+ this is displaying the top 10 routes based on on time performance

```python
# Load the CSV data into a Pandas DataFrame
df = pd.read_csv('on_time.csv')

# Group by routes and calculate the average performance
route_performance = df.groupby('route_full_name')['on_time_percent'].mean().reset_index()

# Sort by performance in descending order and select the top 10 routes
top_10_routes = route_performance.sort_values(by='on_time_percent', ascending=False).head(10)

# Create a horizontal bar chart to visualize the top 10 routes
plt.figure(figsize=(10, 6))
plt.barh(top_10_routes['route_full_name'], top_10_routes['on_time_percent'])
plt.xlabel('Performance')
plt.ylabel('Route')
plt.title('Top 10 Highest Performing Routes')
plt.show()
```
